```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#6 Distributions to be used:
# 1. Binomial
# 2. Gamma
# 3. Poisson
# 4. Geometric
# 5. Beta
# 6. Chi Square
```

```python
#Binomial
binomial = np.random.binomial(1,.5, 100)
plt.hist(binomial, bins=4, color='c')
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean()-binomial.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
print('Descriptive statistics not a strong gauge for this distribution.')
```

```python
#Gamma
gamma = np.random.gamma(shape=5,scale=1, size=100)
plt.hist(gamma, color='c')
plt.axvline(gamma.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() + gamma.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(gamma.mean()-gamma.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
print('Distribution is skewed. Descriptive statistics provide a decent gauge for this distribution in that mean and std deviation seem to be accurate.')
```

```python
#Poisson
poisson = np.random.poisson(3, 100)
plt.hist(poisson, color='c')
plt.axvline(poisson.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean() + poisson.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean()-poisson.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
print('Similar to Gamma distribution above, the Poisson plot is skewed but the descriptive statistics seem to be a relatively good gauge of the distribution.')
```

```python
#Geometric
geometric = np.random.geometric(.5, 100)
plt.hist(geometric, color='c')
plt.axvline(geometric.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(geometric.mean() + geometric.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(geometric.mean()-geometric.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
print('Again, the plot shows skewedness, but the descriptive statistics do not seem to be far off.')
```

```python
#Beta
beta = np.random.beta(1, 5, 100)
plt.hist(beta, color='c')
plt.axvline(beta.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(beta.mean() + beta.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(beta.mean()-beta.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
print('Seems to follow along with the general theme I am seeing in most of these distributions. Skewedness but the descriptive statistics are not a horrible measure of the distribution.')
```

```python
#Chi Square
chi = np.random.chisquare(2, 100)
plt.hist(chi, color='c')
plt.axvline(chi.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(chi.mean() + chi.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(chi.mean()-chi.std(), color='b', linestyle='dashed', linewidth=2) 
plt.show()
print('Distribution is more skewed than the others, descriptive statistics not the best measure of the distribution')
```

```python
#Generate two normally-distributed variables, one with a mean of 5 and 
#standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
#Add them together to create a third variable and then graph that variable using a histogram.
#Compute the mean and standard deviation and plot them as vertical lines on the histogram.
#Evaluate the descriptive statistics against the data.

norm1 = np.random.normal(5, .5, 1000)
norm2 = np.random.normal(10, 1, 1000)
norm3 = norm1 + norm2

#Plot a histogram for norm3
plt.hist(norm3, bins=20, color='c')

# Add a vertical line at the mean.
plt.axvline(norm3.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(norm3.mean() + norm3.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(norm3.mean()-norm3.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()

print('Based on the plot above, the descriptive statistics are fairly strong and this represents a normal distribution for the most part with very slight skew. Most data is within 1 standard deviation of the mean.')
```
