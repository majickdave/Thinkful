### First we will build our Decision Tree Model

```python
#Create a separate sliced dataframe

pokemon_revised = pokemon.iloc[:, 4:11]
```

```python
#Convert class column 'Type 1' to integer. Currently an object, transform to category first, then int.

pokemon['Type 1'].astype('category')
```

```python
# This is the model we'll be using.
from sklearn import tree

# A convenience for displaying visualizations.
from IPython.display import Image

# Packages for rendering our tree.
import pydotplus
import graphviz

# Initialize and train our tree.
decision_tree = tree.DecisionTreeClassifier(
    criterion='entropy',
    max_features=3,
    max_depth=5)

decision_tree.fit(pokemon['Type 1'], pokemon_revised)

# Render our tree.
dot_data = tree.export_graphviz(
    decision_tree, out_file=None,
    feature_names=pokemon_revised.columns,
    class_names=['Grass', 'Fire', 'Water', 'Bug', 'Normal', 'Poison', 'Electric',
       'Ground', 'Fairy', 'Fighting', 'Psychic', 'Rock', 'Ghost', 'Ice',
       'Dragon', 'Dark', 'Steel', 'Flying'],
    filled=True)

graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())

#Print run time
import time

start_time = time.time()
main()
print("--- %s seconds ---" % (time.time() - start_time))
```
