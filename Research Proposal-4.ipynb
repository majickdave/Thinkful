```python
import pandas as pd
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
attrition = pd.read_csv('IBM_Attrition.csv')
```

```python
#Calculate attrition rate by business travel category

attrition_business_travel = attrition[['BusinessTravel', 'Attrition']].groupby('BusinessTravel').agg({'Attrition': 'value_counts'}).add_suffix('_count').reset_index()
attrition_business_travel = pd.merge(attrition_business_travel, attrition_business_travel.groupby('BusinessTravel').sum().reset_index(), on='BusinessTravel')
attrition_business_travel['Attrition Rate'] = attrition_business_travel['Attrition_count_x']/attrition_business_travel['Attrition_count_y']*100

#Confirm this worked
attrition_business_travel
```

```python
# Now plot the attrition rate by business travel category, first create a dataframe with only
# Attrition = 'Yes'

attrition_frequent_travel = attrition_business_travel[attrition_business_travel['Attrition'] == 'Yes']

#Plot
sns.set(style="darkgrid")
g = sns.factorplot(x="BusinessTravel", y="Attrition Rate", data=attrition_frequent_travel,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Attrition Rate")
g.set_xlabels("Business Travel")
plt.title('Attrition Rate by Business Travel Category')
plt.xticks(rotation=45)

plt.show()
```

```python
#Now let's calculate attrition rate by job role

attrition_job_role = attrition[['JobRole', 'Attrition']].groupby('JobRole').agg({'Attrition': 'value_counts'}).add_suffix('_count').reset_index()
attrition_job_role = pd.merge(attrition_job_role, attrition_job_role.groupby('JobRole').sum().reset_index(), on='JobRole')
attrition_job_role['Attrition Rate'] = attrition_job_role['Attrition_count_x']/attrition_job_role['Attrition_count_y']*100

#Confirm this worked
attrition_job_role
```

```python
# Now plot the attrition rate for job role by business travel category, first create a 
#dataframe with only Attrition = 'Yes'

attrition_job_role = attrition_job_role[attrition_job_role['Attrition'] == 'Yes']

#Plot
sns.set(style="darkgrid")
g = sns.factorplot(x="JobRole", y="Attrition Rate", data=attrition_job_role,
                   size=8, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Attrition Rate")
g.set_xlabels("Job Role")
plt.title('Attrition Rate by Job Role')
plt.xticks(rotation=45)
plt.yticks(np.arange(0, 41, 1))

plt.show()
```

```python
#Based on plots above, filter dataframe based on Job Role = 'Sales Representative'
attrition_salesrep = attrition[attrition['JobRole'] == 'Sales Representative']
```

```python
#Calculate attrition rate for sales representatives by business travel category
attrition_salesrep_attrition = attrition_salesrep[['BusinessTravel', 'Attrition']].groupby('BusinessTravel').agg({'Attrition': 'value_counts'}).add_suffix('_count').reset_index()
attrition_salesrep_attrition = pd.merge(attrition_salesrep_attrition, attrition_salesrep_attrition.groupby('BusinessTravel').sum().reset_index(), on='BusinessTravel')
attrition_salesrep_attrition['Attrition Rate'] = attrition_salesrep_attrition['Attrition_count_x']/attrition_salesrep_attrition['Attrition_count_y']*100

#Confirm this worked
attrition_salesrep_attrition
```

```python
# Now plot the attrition rate for sales representatives by business travel category, 
#first create a dataframe with only Attrition = 'Yes'

attrition_salesrep_yes = attrition_salesrep_attrition[attrition_salesrep_attrition['Attrition'] == 'Yes']

#Plot
sns.set(style="darkgrid")
g = sns.factorplot(x="BusinessTravel", y="Attrition Rate", data=attrition_salesrep_yes,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Attrition Rate")
g.set_xlabels("Business Travel")
plt.title('Attrition Rate for Sales Represenatives by Business Travel Category')
plt.xticks(rotation=45)
plt.yticks(np.arange(0, 70, 2))

plt.show()
```

# Experiment Set Up

## Two Versions: 
Control - Random selection of sales representatives who currently have long term frequent travel based assignment.

Treatment - Random assignment of non-travel based assignment for Sales Representatives who previously had frequent travel based assignment.

## Sample:
Treatment will be randomly applied to 10 employees and the control will be a random selection of 10 employees.

## Hypothesis:
Reduction in travel will lower the overall numbers of attrition.

## Outcome:
Key Metric - Attrition

## Other Measured Variables:
Secondary Metrics - job, environment, relationship, and work life balance satisfaction

___________________________________________________________________________________________________

## Setting the Stage: 
Based on the my findings from the exploratory analysis above, it seems that attrition rates are higher for those employees who both travel frequently, and hold the job role of sales representative. Further, within the category of sales representatives, those who did not travel had no attrition recorded. IBM wants to uncover why attrition may be higher for these particular conditions and have asked their data science team to conduct an experiment to potentially causality. The experiment to be conducted will be randomly assigning non-travel based work assignments (treatment) for 10 sales representatives coming off a frequent travel based assignment over the next year and randomly selecting 10 employees who currently hold a long term, frequent travel based work assignment (control) during the same time frame.

## Rollout Plan: 
The participating inidividuals selected for the treatment group will be randomly chosen based on the end of their current assignment. The control group can be randomly selected among the population of sales representatives who have a long term, frequent travel based assignment. Based on this information, the rollout is classified as gradual. The intention is to "treat" 10 sales representatives with this new work assignment and the duration of the project should be 1 year from when the new assignment began. The reason for this gradual rollout is not only based on the fact that sales representatives have current ongoing obligations, but also due to the fact that if for some reason there are immediate negative consequences, the experiment can be paused or ended.

## Evaluation Plan:
### Key Metric:
For our key metric of attrition, we will evaluate success based on a p-value of .05 in regard to our two samples. To be clear, we will measure attrition of these two groups over the course of the year, from the time the randomly assigned, non-travel based work assignment occurred. We will refer to this average as attrition rate.

### Secondary Metrics:
For our secondary metrics, we will also measure based on a p-value of .05 in regard to our two samples. Toward the end of this assignment, our treatment and control groups will take an employee survey simliar to what spawned the experiment. We will compare the variables of job, environment, relationship, and work life balance satisfaction between the two groups.

## Infrastructure Setup:
We will also ensure that the test is prepared by coordinating with HR and managers to ensure that the job assignments fall in line with the employees' areas of expertise. This will help ensure continuity and randomness of the experiment itself.
