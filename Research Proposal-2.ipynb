```python
# Boxplot of monthly income by job role, want to specifically see how
#Sales representative fits in.

plt.figure(figsize=[17,5])
ax = sns.boxplot(x='JobRole',y='MonthlyIncome', data=attrition)  
plt.title('Boxplot of Job Role vs. Monthly Income', fontsize=20)
sns.despine(offset=10, trim=True)
ax.set(xlabel='Job Role', ylabel='Monthly Income')
plt.xticks(rotation=45)
plt.show()
```

```python
#Based on plots above, filter dataframe based on Job Role = 'Sales Representative'
attrition_findings = attrition[attrition['JobRole'] == 'Sales Representative']
```

```python
#Now look at age and income of this filtered group using a scatter plot with a line
#to understand the characteristics of this smaller group

g = sns.lmplot(y='MonthlyIncome', # Variable 1.
               x='Age', # Variable 2.
               data=attrition_findings, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("Monthly Income")
g.set_xlabels("Employee Age")
plt.title('Employee Age vs. Monthly Income')
plt.show()
```

Experiment Set Up

Two Versions: 
Control - Random assignment of frequent travel based assignment for Sales Representatives.
Treatment - Random assignment of non-travel based assignment for Sales  Representatives.

Sample - Amongst those with Sales Representative job role, the treatment will be randomly applied to 10 employees (non-travel assignment) and the control will also be randomly applied to 10 employees (frequent travel assignment).

Hypothesis:
Reduction in travel (non-travel assignment) will lower the overall numbers of attrition.

Outcome:
Key Metric - Attrition

Other Measured Variables:
Secondary Metrics - job, environment, relationship, and work life balance satisfaction

___________________________________________________________________________________________________________________________________

Setting the Stage: Based on the my findings from the exploratory analysis above, it seems that attrition rates are higher for those employees who both travel frequently, and hold the job role of sales representative. IBM wants to uncover why attrition may be higher for these particular conditions and have asked their data science team to conduct an experiment to potentially causality. The experiment to be conducted will be randomly assigning non-travel based work assignments (treatment) for 10 sales representatives over the next year and randomly assigning frequent travel based work assignments (control) to 10 employees during the same time frame.

Rollout Plan: The participating inidividuals selected for the treatment and control groups would be randomly chosen based on the end of their current assignment. Based on this information, the rollout is classified as gradual. The intention is to "treat" 10 sales representatives with this new work assignment and the duration of the project should be 1 year from when the new assignment began. The reason for this gradual rollout is not only based on the fact that sales representatives have current ongoing obligations, but also due to the fact that if for some reason there are immediate negative consequences, the experiment can be paused or ended.

Evaluation Plan: 
Key Metric - For our key metric of attrition, we will evaluate success based on a p-value of .05 in regard to our two samples. To be clear, we will measure attrition of these two groups over the course of the year, from the time the randomly assigned, non-travel based work assignment occurred. We will refer to this average as attrition rate.

Secondary Metrics - For our secondary metrics, we will also measure based on a p-value of .05 in regard to our two samples. Toward the end of this assignment, our treatment and control groups will take an employee survey simliar to what spawned the experiment. We will compare the variables of job, environment, relationship, and work life balance satisfaction between the two groups.

Infrastructure Setup:
We will also ensure that the test is prepared by coordinating with HR and managers to ensure that the job assignments fall in line with the employees' areas of expertise. This will help ensure continuity and randomness of the experiment itself.
